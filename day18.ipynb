### Graph Functions
1. plot                     - simple line plot
2. plot.area                - Filled version of line plot
3. plot.bar /plot.barh      - bar graph
4. plot.box                 - Box plot
5. plot.density             - Density line(Smooth curve)
6. plot.hexbin              - Hexagon Version of Scatter with density
7. plot.hist                - histogram
8. plot.kde                 - kernel density estimator(Density graph)
9. plot.pie                 - pie chart of categorical data
10. plot.scatter            - scatter graph

# Seaborn

```python
import seaborn as sns
```

### Categories
1. Categorical plots (hue can be used | to compare with categorical column)
     a. Countplot
     b. Barplot (categorical and numerical data required each| calculates mean by default)
     c. Box plot
     d. Violin plot
     e. swarm plot
2. Distribution plots
     a. Distplot
     b. Displot
     c. Jointplot
     d. Pairplot
     e. Kdeplot
     f. Rugplot
     g. Scatterplot
3. Matrix plots
     a. Heatmap
     b. Clustermap
     c. Pairplot(same as above one)
4. Regression plots
     a. Lmplot
     b. Regplot
5. Grid plots
     a. Pairgrid
     b. Jointgrid
     c. Facetgrid


```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.countplot(x='tip',data=tips , hue='sex')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=sum)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='time' , estimator=np.mean)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time' )
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='time' )
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='time' )
```
