```python
import numpy as np
import pandas as pd
```

```python
data = {'Country':['India','USA','Japan','China','India','USA'],
       'Year':['2010','2010','2010','2015','2015','2015'],
       'Continent':['Asia','NA','Asia','Asia','Asia','NA'],
       'Value':[20,10,40,50,60,55]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
pd.pivot_table(df,index=['Year'],aggfunc='sum')
```

```python
pd.pivot_table(df,index=['Continent'],aggfunc='sum')
```

## CONCAT

```python
people1 = {'id':[1,2,3,4],
           'name':['ab','bc','cd','df'],
           'age':[45,25,21,50]
          }

people2 = {'id':[11,22,33,44],
          'name':['qwe','asd','zxc','tuh'],
          'age':[34,56,78,90]
          }

people3 = {'id':[111,222,333,444,],
          'name':['rty','dfg','cvb','ujm'],
          'age':[45,67,90,43]
          }
```

```python
df1=pd.DataFrame(people1)
df2=pd.DataFrame(people2)
df3=pd.DataFrame(people3)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
people4 = {'id':[1,2,3,4],
           'name':['ab','bc','cd','df'],
           'sales':[56,89,54,23]
          }

people5 = {'id':[1,2,3,4],
          'name':['ab','bc','cd','df'],
          'age':[344,565,785,906]
          }
```

```python
df4=pd.DataFrame(people4)
df5=pd.DataFrame(people5)
```

```python
df4
```

```python
df5
```

```python
pd.merge(df4,df5,on=['id','name'])
```

```python
people6 = {'id':[1,2,3,4],
           'name':['ab','bc','cd','df'],
           'sales':[56,89,54,23]
          }

people7 = {
          's_name':['ab','bc','cd','df'],
          'ph':[344,565,785,906]
          }
```

```python
df6 = pd.DataFrame(people6)
df7 = pd.DataFrame(people7)
```

```python
df6
```

```python
df7
```

```python
df6.join(df7)
```
